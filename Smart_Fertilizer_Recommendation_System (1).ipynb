```python
import pandas as pd

# Load crop dataset
df_crop = pd.read_csv('Crop_recommendation.csv')
print("Crop Dataset Loaded")
print(df_crop.head())

# Load fertilizer dataset
df_fert = pd.read_csv('Dataset.csv')
print("\nFertilizer Dataset Loaded")
print(df_fert.head())
```

```python
print("Crop Dataset Shape:", df_crop.shape)
print("Crop Dataset Columns:", df_crop.columns)
df_crop.head()
```

```python
X_crop = df_crop.drop('label', axis=1)
y_crop = df_crop['label']

X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(X_crop, y_crop, test_size=0.2, random_state=42)

model_crop = RandomForestClassifier()
model_crop.fit(X_train_c, y_train_c)

y_pred_c = model_crop.predict(X_test_c)

print("Crop Model Accuracy:", accuracy_score(y_test_c, y_pred_c))
print(classification_report(y_test_c, y_pred_c))
```

```python
def recommend_crop(n, p, k, temp, humidity, ph, rainfall):
    data = np.array([[n, p, k, temp, humidity, ph, rainfall]])
    result = model_crop.predict(data)
    return result[0]

# Example
recommend_crop(90, 42, 43, 20.5, 80.0, 6.5, 200.0)
```

```python
from sklearn.preprocessing import LabelEncoder

le_soil = LabelEncoder()
le_crop = LabelEncoder()
le_fert = LabelEncoder()

df_fert['Soil'] = le_soil.fit_transform(df_fert['Soil'])
df_fert['Crop'] = le_crop.fit_transform(df_fert['Crop'])
df_fert['Fertilizer'] = le_fert.fit_transform(df_fert['Fertilizer'])
```

```python
X_fert = df_fert.drop('Fertilizer', axis=1)
y_fert = df_fert['Fertilizer']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

X_train_f, X_test_f, y_train_f, y_test_f = train_test_split(X_fert, y_fert, test_size=0.2, random_state=42)

model_fert = RandomForestClassifier()
model_fert.fit(X_train_f, y_train_f)

y_pred_f = model_fert.predict(X_test_f)

print("Fertilizer Model Accuracy:", accuracy_score(y_test_f, y_pred_f))
print(classification_report(y_test_f, y_pred_f))
```

```python
print("Soil types:", le_soil.classes_)
print("Crop types:", le_crop.classes_)
```

```python
import pandas as pd
import numpy as np

def recommend_fertilizer(temp, humidity, rainfall, ph, n, p, k, soil, crop):
    if soil not in le_soil.classes_:
        return f"Soil type '{soil}' not found in training data"
    if crop not in le_crop.classes_:
        return f"Crop type '{crop}' not found in training data"
    soil_encoded = le_soil.transform([soil])[0]
    crop_encoded = le_crop.transform([crop])[0]

    input_dict = {
        'Temperature': [temp],
        'Humidity': [humidity],
        'Rainfall': [rainfall],
        'pH': [ph],
        'N': [n],
        'P': [p],
        'K': [k],
        'Soil': [soil_encoded],
        'Crop': [crop_encoded]
    }

    input_df = pd.DataFrame(input_dict)
    pred_encoded = model_fert.predict(input_df)
    fertilizer_name = le_fert.inverse_transform(pred_encoded)
    return fertilizer_name[0]
```

```python
recommend_fertilizer(30, 80, 200, 6.5, 40, 20, 30, 'alluvial', 'rice')
```

```python
import ipywidgets as widgets
from IPython.display import display, HTML

# Dropdown widgets for soil and crop (use your actual classes)
soil_dropdown = widgets.Dropdown(
    options=list(le_soil.classes_),
    description='Soil:',
)

crop_dropdown = widgets.Dropdown(
    options=list(le_crop.classes_),
    description='Crop:',
)

# Numeric inputs for other features
temp_input = widgets.FloatText(description='Temperature:')
humidity_input = widgets.FloatText(description='Humidity:')
rainfall_input = widgets.FloatText(description='Rainfall:')
ph_input = widgets.FloatText(description='pH:')
n_input = widgets.IntText(description='N:')
p_input = widgets.IntText(description='P:')
k_input = widgets.IntText(description='K:')

# Button to run recommendation
button = widgets.Button(description="Recommend Fertilizer")

# Output area
output = widgets.Output()

def on_button_clicked(b):
    with output:
        output.clear_output()
        # Read all inputs
        temp = temp_input.value
        humidity = humidity_input.value
        rainfall = rainfall_input.value
        ph = ph_input.value
        n = n_input.value
        p = p_input.value
        k = k_input.value
        soil = soil_dropdown.value
        crop = crop_dropdown.value

        # Call recommendation function
        result = recommend_fertilizer(temp, humidity, rainfall, ph, n, p, k, soil, crop)
        print("Recommended Fertilizer:", result)

button.on_click(on_button_clicked)

# Display title
display(HTML("<h1 style='color: green; text-align: center;'>Smart Fertilizer Recommendation System</h1>"))

# Display form widgets
display(temp_input, humidity_input, rainfall_input, ph_input, n_input, p_input, k_input, soil_dropdown, crop_dropdown, button, output)
```
